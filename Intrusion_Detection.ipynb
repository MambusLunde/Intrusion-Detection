```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.svm import SVC
from xgboost.sklearn import XGBClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
dataset = pd.read_csv('https://media.githubusercontent.com/media/MambusLunde/Intrusion-Detection/main/Hikari2021.csv')
```

```python
le = LabelEncoder()
labels = ['Benign','Background','Probing','Bruteforce','Bruteforce-XML','XMRIGCC CryptoMiner']
le.fit(labels)
dataset['EncodedLabel'] = le.transform(dataset.traffic_category)
```

```python
dataset.drop(axis=0,columns=['Unnamed: 0','Unnamed: 0.1','uid','originh','responh','traffic_category'],inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset.loc[:,~dataset.columns.isin(['Label','EncodedLabel'])], 
                                                    dataset.loc[:,dataset.columns.isin(['Label','EncodedLabel'])], 
                                                    test_size=0.2,
                                                    random_state=28,
                                                    stratify=dataset.EncodedLabel)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
ct = ColumnTransformer([
    ('standard scaler', StandardScaler(), X_train.columns[2:])],
    remainder='passthrough'
)
X_train = ct.fit_transform(X_train)
X_test = ct.transform(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=6, n_jobs=-1)
knn = knn.fit(X_train, y_train['EncodedLabel'])
train_score = knn.score(X_train[:1000], y_train['EncodedLabel'][:1000])
test_score = knn.score(X_test[:1000], y_test['EncodedLabel'][:1000])
print(f"KNN Train Accuracy: {train_score}")
print(f"KNN Test Accuracy: {test_score}")
```

```python
svm = SVC()
svm = svm.fit(X_train, y_train['EncodedLabel'])
train_score = svm.score(X_train[:1000], y_train['EncodedLabel'][:1000])
test_score = svm.score(X_test[:1000], y_test['EncodedLabel'][:1000])
print(f"svm Train Accuracy: {train_score}")
print(f"svm Test Accuracy: {test_score}")
```

```python
XGB = XGBClassifier()
XGB = XGB.fit(X_train, y_train['EncodedLabel'])
train_score = XGB.score(X_train[:1000], y_train['EncodedLabel'][:1000])
test_score = XGB.score(X_test[:1000], y_test['EncodedLabel'][:1000])
print(f"XGB Train Accuracy: {train_score}")
print(f"XGB Test Accuracy: {test_score}")
```

```python
SGD = SGDClassifier()
SGD = SGD.fit(X_train, y_train['EncodedLabel'])
train_score = SGD.score(X_train[:1000], y_train['EncodedLabel'][:1000])
test_score = SGD.score(X_test[:1000], y_test['EncodedLabel'][:1000])
print(f"SGD Train Accuracy: {train_score}")
print(f"SGD Test Accuracy: {test_score}")
```

```python
RFC = RandomForestClassifier()
RFC = RFC.fit(X_train, y_train['EncodedLabel'])
train_score = RFC.score(X_train[:1000], y_train['EncodedLabel'][:1000])
test_score = RFC.score(X_test[:1000], y_test['EncodedLabel'][:1000])
print(f"RFC Train Accuracy: {train_score}")
print(f"RFC Test Accuracy: {test_score}")
```

```python
GBC = GradientBoostingClassifier()
GBC = GBC.fit(X_train, y_train['EncodedLabel'])
train_score = GBC.score(X_train[:1000], y_train['EncodedLabel'][:1000])
test_score = GBC.score(X_test[:1000], y_test['EncodedLabel'][:1000])
print(f"GBC Train Accuracy: {train_score}")
print(f"GBC Test Accuracy: {test_score}")
```
